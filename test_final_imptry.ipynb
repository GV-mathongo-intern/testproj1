```python
import pandas as pd
import json
```

```python
with open('strapi-admin-panel.detailed_analyses_small.json') as outf:
    df=json.load(outf)
```

```python
df1=pd.DataFrame.from_dict(df)
df1.head()
```

```python
df1['_id1']=df1['_id'].apply(lambda x:x.get('$oid'))
df1['testid1']=df1['testId'].apply(lambda x:x.get('$oid'))
df1['userid1']=df1['userId'].apply(lambda x:x.get('$oid'))
df1.head()
```

```python
df1.drop(['_id','userId','testId'],axis=1,inplace=True)
df1.head()
```

```python
df2=pd.DataFrame([tuple(list(d.values()))+(tup._9,) for tup in df1.itertuples() for d in tup.subjects])
df2.columns=list(df1['subjects'][0][0].keys())+['id1']
df2.head()
```

```python
for col in df2.columns:
    print(col)
```

```python
df3=df2.drop(['painfulQuestions','missedConcepts','_id','subjectId'],axis=1)
df3.head()
```

```python
df4=df3.groupby('id1').aggregate('sum')
df4.head()
```

```python
df4['accuracy']=(df4['totalCorrect']/df4['totalAttempted'])*100
df4.head()
```

```python
df4['easyCorrect1']=(df4['easyCorrect']/(df4['easyCorrect']+df4['easyIncorrect']+df4['easyUnanswered']))*100
df4['easyIncorrect1']=(df4['easyIncorrect']/(df4['easyCorrect']+df4['easyIncorrect']+df4['easyUnanswered']))*100
df4['easyUnanswered1']=(df4['easyUnanswered']/(df4['easyCorrect']+df4['easyIncorrect']+df4['easyUnanswered']))*100
df4['moderateCorrect1']=(df4['moderateCorrect']/(df4['moderateCorrect']+df4['moderateIncorrect']+df4['moderateUnanswered']))*100
df4['moderateIncorrect1']=(df4['moderateIncorrect']/(df4['moderateCorrect']+df4['moderateIncorrect']+df4['moderateUnanswered']))*100
df4['moderateUnanswered1']=(df4['moderateUnanswered']/(df4['moderateCorrect']+df4['moderateIncorrect']+df4['moderateUnanswered']))*100
df4['toughCorrect1']=(df4['toughCorrect']/(df4['toughCorrect']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4['toughIncorrect1']=(df4['toughIncorrect']/(df4['toughCorrect']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4['toughUnanswered1']=(df4['toughUnanswered']/(df4['toughCorrect']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4.head()
```

```python
df5=df4.drop(columns=['easyCorrect','easyIncorrect','easyUnanswered','moderateCorrect','moderateIncorrect','moderateUnanswered','toughCorrect','toughIncorrect','toughUnanswered','notVisited','totalAttempted','totalCorrect'],axis=1)
df5.head()
```

```python
df5['accuracy']=df5['accuracy'].fillna(-1)
df5.head()
```

```python
df4['notVisited1']=(df4['notVisited']/(df4['totalCorrect']+df4['easyIncorrect']+df4['easyUnanswered']+df4['moderateIncorrect']+df4['moderateUnanswered']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4['totalAttempted1']=(df4['totalAttempted']/(df4['totalCorrect']+df4['easyIncorrect']+df4['easyUnanswered']+df4['moderateIncorrect']+df4['moderateUnanswered']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4['totalCorrect1']=(df4['totalCorrect']/(df4['totalCorrect']+df4['easyIncorrect']+df4['easyUnanswered']+df4['moderateIncorrect']+df4['moderateUnanswered']+df4['toughIncorrect']+df4['toughUnanswered']))*100
df4.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df6_1=sc.fit_transform(df6)
pd.DataFrame(df6_1).describe()
```

```python
a=df6.values.tolist()
```

```python
a[0]
```

```python
from sklearn.cluster import OPTICS,KMeans
m=OPTICS(min_samples=12).fit(df6_1)
n=KMeans(6).fit(df6_1)
```

```python
m.get_params()
```

```python
set(n.labels_)
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

```

```python
silhouette = silhouette_score(a, m.labels_)
db_index = davies_bouldin_score(a, m.labels_)
ch_index = calinski_harabasz_score(a, m.labels_)
silhouette1 = silhouette_score(a, n.labels_)
db_index1 = davies_bouldin_score(a, n.labels_)
ch_index1 = calinski_harabasz_score(a, n.labels_)
print(f"Silhouette Score: {silhouette:.2f},{silhouette1:.2f}")
print(f"Davies-Bouldin Index: {db_index:.2f},{db_index1:.2f}")
print(f"Calinski-Harabasz Index: {ch_index:.2f},{ch_index:.2f}")
```

```python
import matplotlib.pyplot as plt
SSE = []
for cluster in range(1,20):
    kmeans = KMeans( n_clusters = cluster, init='k-means++')
    kmeans.fit(df6_1)
    SSE.append(kmeans.inertia_)

frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```
