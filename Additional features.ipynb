```python

params = {
    "n_neighbors": n_neighbors,
    "n_components": n_components,
    "eigen_solver": "auto",
    "random_state": 0,
}


lle_mod = manifold.LocallyLinearEmbedding(method="modified", **params)
S_standard = lle_mod.fit_transform(df3_0)

S_standard

kmeans = KMeans()
ssd = []
K = range(1, 30)

for k in K:
    kmeans = KMeans(n_clusters=k , n_init=10).fit(S_standard)
    ssd.append(kmeans.inertia_)

ssd

# plt.plot(K, ssd, "bx-")
# plt.xlabel("Distance Residual Sums for K Values (WCSS)")
# plt.title("Elbow Method for Optimum Number of Clusters")
# plt.show()

kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k=(2, 20))
visu.fit(S_standard)
visu.show()

```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42).fit(S_standard)
clusters = kmeans.labels_

pd.DataFrame({"Customers": df3_0.index, "Clusters": clusters})
df3_0["cluster_no"] = clusters
df3_0.head()


silhouette = silhouette_score(df3_0.drop(columns=['cluster_no']), kmeans.labels_)
db_index = davies_bouldin_score(df3_0.drop(columns=['cluster_no']), kmeans.labels_)
ch_index = calinski_harabasz_score(df3_0.drop(columns=['cluster_no']), kmeans.labels_)
print(silhouette,db_index,ch_index)
```
