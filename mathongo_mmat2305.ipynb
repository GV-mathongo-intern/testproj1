### 2.5 Test-Wise

`Here on I wanted to better understand how each exam progressed`

` I first gathered data regarding how much time was spent and when in the course of the exam was that amount if time spent, in which section was the question in, and did he answer it correctly or incorrectly or did he not answer it `

```python
his1=[]
c=0
for i in df['sections'][0]:
    c=c+1
    for j in i.get('questions'):
        if len(j.get('markedOptions'))==0:
            d=-1
        elif j.get('markedOptions')[0].get('isCorrect') is True:
            d=1
        else:
            d=0
        his1.append([j.get('timeTaken'),j.get('timeLeftWhenAttempted'),c,d,j.get('questionId').get('$oid')])
```

`Using the gathered data I have created a dataframe. Now when you sort this dataframe in the descending order of the column timeLeftWhenAttempted we will have in our hands the entire timeline if the exam`

```python
df3=pd.DataFrame(his1,columns=['timeTaken','timeLeftWhenAttempted','section','correct','q_id'])
df3=df3.sort_values(by=['timeLeftWhenAttempted'],ascending=False)
df3.head()
```
