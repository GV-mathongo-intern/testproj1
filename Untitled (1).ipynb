```python
from sklearn.decomposition import FastICA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_components_range = range(1, 13)  # Range of components from 1 to 12
```

```python
silhouette_scores = []
distortions = []
```

```python
for n_components in n_components_range:
    # Transform data using ICA
    ica = FastICA(n_components=n_components)
    data_transformed = ica.fit_transform(adf8_arr)

    # Determine number of clusters using elbow method
    distortions.append(sum(np.min(np.square(adf8_arr - ica.inverse_transform(data_transformed)),axis=1)))
    kmeans = KMeans(n_clusters=4).fit(data_transformed)
    silhouette_scores.append(silhouette_score(adf8_arr, kmeans.labels_))

```
